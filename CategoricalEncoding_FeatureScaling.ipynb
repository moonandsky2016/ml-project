```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, RobustScaler

# Create DataFrame
data = {
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female'],
    'City': ['Mumbai', 'Delhi', 'Chennai', 'Mumbai', 'Delhi'],
    'Size': ['Small', 'Large', 'Medium', 'Large', 'Small'],
    'Age': [25, 45, 32, 60, 28],
    'Fare': [15, 300, 85, 450, 20]
}

df = pd.DataFrame(data)

# -------------------------------
# Task 1 — Categorical Encoding
# -------------------------------

# 1. Gender → Label Encoding
le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])  # Female=0, Male=1

# 2. City → One-Hot Encoding (drop one column)
df = pd.get_dummies(df, columns=['City'], drop_first=True)

# 3. Size → Custom Label Encoding (Ordered)
size_mapping = {'Small': 0, 'Medium': 1, 'Large': 2}
df['Size'] = df['Size'].map(size_mapping)

print("After Encoding:\n", df)

# -------------------------------
# Task 2 — Feature Scaling
# -------------------------------

# RobustScaler is used because Fare has outliers (e.g., 450)
# It uses median and IQR, making it less sensitive to extreme values

scaler = RobustScaler()

df[['Age', 'Fare']] = scaler.fit_transform(df[['Age', 'Fare']])

print("\nAfter Scaling:\n", df)
```
